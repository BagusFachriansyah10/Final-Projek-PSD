FIR HAMMING WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter lowpass FIR
def design_lowpass_fir(numtaps, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    taps = signal.firwin(numtaps, norm_cutoff, window='hamming', pass_zero=True)
    return taps

# Fungsi untuk membuat filter highpass FIR
def design_highpass_fir(numtaps, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    taps = signal.firwin(numtaps, norm_cutoff, window='hamming', pass_zero=False)
    return taps

# Baca file WAV
input_wav_path = '060591718.wav'
output_wav_path = 'output_FIR_hamming_bandpass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter low pass
numtaps_lowpass = 101  # Panjang filter low pass
cutoff_lowpass = 1500  # Frekuensi cutoff low pass dalam Hz
taps_lowpass = design_lowpass_fir(numtaps_lowpass, cutoff_lowpass, fs)

# Desain filter high pass
numtaps_highpass = 101  # Panjang filter high pass
cutoff_highpass = 500  # Frekuensi cutoff high pass dalam Hz
taps_highpass = design_highpass_fir(numtaps_highpass, cutoff_highpass, fs)

# Gabungkan kedua filter dengan konvolusi
taps_bandpass = np.convolve(taps_lowpass, taps_highpass, mode='same')

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps_bandpass, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps_bandpass)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps_bandpass, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

FIR KAISER WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass FIR
def design_bandpass_fir(numtaps, f_low, f_high, fs):
    nyquist = 0.5 * fs
    norm_low = f_low / nyquist
    norm_high = f_high / nyquist
    taps = signal.firwin(numtaps, [norm_low, norm_high], pass_zero=False, window=('kaiser', 8.0))
    return taps

# Baca file WAV
input_wav_path = '060591718.wav'
output_wav_path = 'output_FIR_kaiser_bandpass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter bandpass
numtaps_bandpass = 101  # Panjang filter bandpass
f_low = 500  # Frekuensi cutoff rendah dalam Hz
f_high = 1500  # Frekuensi cutoff tinggi dalam Hz
taps_bandpass = design_bandpass_fir(numtaps_bandpass, f_low, f_high, fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps_bandpass, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps_bandpass)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps_bandpass, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5*fs*w/np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

IIR CHEBYSHEV WINDOW

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter bandpass IIR dengan jendela Chebyshev
def design_bandpass_iir(order, lowcut, highcut, fs, ripple):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = signal.cheby1(order, ripple, [low, high], btype='band', analog=False)
    return b, a

# Baca file WAV
input_wav_path = '060591718.wav'
output_wav_path = 'output_IIR_chebyshev_bandpass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter bandpass IIR dengan jendela Chebyshev
order_bandpass = 4  # Orde filter band pass
lowcut_bandpass = 500  # Frekuensi cutoff low pass dalam Hz
highcut_bandpass = 1500  # Frekuensi cutoff high pass dalam Hz
ripple = 5  # Ripple (dB) untuk filter Chebyshev
b, a = design_bandpass_iir(order_bandpass, lowcut_bandpass, highcut_bandpass, fs, ripple)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 3)
plt.plot(0.5 * fs * w / np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

# Plot phase response
plt.subplot(4, 1, 4)
plt.plot(0.5 * fs * w / np.pi, np.angle(h))
plt.title('Phase Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase')

plt.tight_layout()
plt.show()
```
